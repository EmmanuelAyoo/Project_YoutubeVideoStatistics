```python
# Dependencies
import requests
import json

# Google API Key
from config import g_key
```

```python
# Build URL using Google Maps API 
base_url = "https://www.googleapis.com/youtube/v3/videoCategories?part=snippet&regionCode=us&key="+g_key
```

```python
US_data = requests.get(base_url).json()
print(json.dumps(US_data, indent=4, sort_keys=True))
```

```python
# Print id and title for the first video
print(US_data["items"][0]["id"])
print(US_data["items"][0]["snippet"]["title"])
```

```python
categories = US_data["items"]

for category in categories:
    print(category["id"],category["snippet"]["title"])
```
