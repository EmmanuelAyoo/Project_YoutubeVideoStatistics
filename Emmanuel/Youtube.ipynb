```python
# Dependencies
import os
import csv
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Save path to data set in a variable
data_file = "Resources/USvideos.csv"
```

```python
#Read the CSV file to get the Data
data_file_df = pd.read_csv(data_file)
data_file_df.head(3)
```

```python
#View the column series in the data
data_file_df = pd.read_csv(data_file)
data_file_df.columns
```

```python
#Select Column Headers that are vital for our analysis
organized_df = data_file_df[[ "category_id", "views", "likes", "dislikes", "comment_count"]] 
organized_df.head()
```

```python
#Review count to ensure no null values exist in the data
organized_df.count()
```

```python
#Apply groupby fxn to compare category_id vs views
grp_organized_df = organized_df.groupby(['category_id'])

views_df = pd.DataFrame(grp_organized_df['views'].sum())

views_df = views_df.sort_values(by =["views"], ascending=False)

views_df.head()
```

```python
# Use DataFrame.plot() in order to create a bar chart of the data
views_df.plot(kind="bar", figsize=(5,3))

# Set a title for the chart
plt.title("Views Per Category_ID")

plt.show()
plt.tight_layout()
```

```python
#Apply groupby fxn to compare category_id vs likes
grp_organized_df = organized_df.groupby(['category_id'])

likes_df = pd.DataFrame(grp_organized_df['likes'].sum())

likes_df = likes_df.sort_values(by =["likes"], ascending=False)

likes_df.head()
```

```python
# Use DataFrame.plot() in order to create a bar chart of the data
likes_df.plot(kind="bar", figsize=(5,3))

# Set a title for the chart
plt.title("Likes Per Category_ID")

plt.show()
plt.tight_layout()
```

```python
#Apply groupby fxn to compare category_id vs dislikes
grp_organized_df = organized_df.groupby(['category_id'])

dislikes_df = pd.DataFrame(grp_organized_df['dislikes'].sum())

dislikes_df = dislikes_df.sort_values(by =["dislikes"], ascending=False)

dislikes_df.head()
```

```python
# Use DataFrame.plot() in order to create a bar chart of the data
dislikes_df.plot(kind="bar", figsize=(5,3))

# Set a title for the chart
plt.title("Dislikes Per Category_ID")

plt.show()
plt.tight_layout()
```

```python
#Apply groupby fxn to compare category_id vs comment_count
grp_organized_df = organized_df.groupby(['category_id'])

comment_count_df = pd.DataFrame(grp_organized_df['comment_count'].sum())

comment_count_df = comment_count_df.sort_values(by =["comment_count"], ascending=False)

comment_count_df.head()
```

```python
# Use DataFrame.plot() in order to create a bar chart of the data
comment_count_df.plot(kind="bar", figsize=(5,3))

# Set a title for the chart
plt.title("Comment Count Per Category_ID")

plt.show()
plt.tight_layout()
```
