```python
# Import Dependencies

import pandas as pd

from datetime import datetime
```

```python
# Reference the CSV file

file = "USvideos.csv"

#Read the CSV's into a Pandas Dataframe

videos_df = pd.read_csv(file)

#Print header

videos_df.head()
```

```python
#Pull out the columns I want to expolore

organized_videos_df = videos_df[["title", "video_id", "trending_date", "publish_time", "views"]]

organized_videos_df.head()
```

```python
#Rename

renamed_df = organized_videos_df.rename(columns={"title":"Video Title",
                                                 "video_id":"Video ID",
                                                "trending_date":"Trending Date",
                                                "publish_time":"Time Published",
                                                "views":"Total Views",})

renamed_df.head()
```

```python
#Look for rows with missing data

renamed_df.count()
```

```python

#Look for duplicates

duplicated_df = renamed_df[renamed_df.duplicated()]

duplicated_df
```

```python
len(duplicated_df)
```

```python
duplicated_df.head()
```

```python
#Delete duplicates

cleanedyoutube_df = renamed_df.drop_duplicates(subset=None, keep="first", inplace=False)

#Check to see that the duplicates are gone

len(cleanedyoutube_df)
```

```python
cleanedyoutube_df.head()
```

```python
#Export cleaned data to a new csv
cleanedyoutube_df.to_csv("CleanedYouTube.csv", index=True, header=True)
```
